```python
import numpy as np
import time
import os
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta
```

```python
def get_50nsedata():
    data = pd.read_csv('ind_nifty50.csv')
    nse_50data = data['Symbol'].to_list()
    nse_50data = nse_50data[1:]
    return nse_50data
```

```python
def get_200nsedata():
    data = pd.read_csv('ind_nifty200.csv')
    nse_200data = data['Symbol'].to_list()
    return nse_200data
```

```python
def get_histdata(ticker, startdate, enddate, interval,Nse_data):
    tickk = ["{}.ns".format(tick) for tick in ticker]
    for line in tickk:
        symbol = line.split(",")[0]
        data = yf.download(symbol, start=startdate, end=enddate, interval=interval)
        if Nse_data == 'Nifty50':
            data.to_csv('C:/Users/admin/Desktop/option chain analysis/datasets/daily/{}.csv'.format(symbol))
        elif Nse_data == 'Nifty200':
            data.to_csv('C:/Users/admin/Desktop/option chain analysis/datasets/monthly/{}.csv'.format(symbol))
        else:
            print('do nothing')
    return data
```

```python
def get_volume_analysis(Nse_data):
    
    output = []
    if Nse_data == 'Nifty50':
        for filename in os.listdir('datasets/daily'):
            week52 = pd.read_csv('datasets/daily/{}'.format(filename))
            week52 = week52[(week52[['Volume']] != 0).all(axis=1)]
            name = filename.replace('.csv','')
            week52.to_csv('C:/Users/admin/Desktop/option chain analysis/datasets/daily/{}.csv'.format(name))
            prev_avg = week52['Volume'].iloc[0:-1].mean()
            last_traded_price = week52['Volume'].iloc[-1]
            prev_avg = int(prev_avg)
            last_traded_price = last_traded_price.item()
            if last_traded_price > prev_avg:
                output.append(filename)
    elif Nse_data == 'Nifty200':
        for filename in os.listdir('datasets/monthly'):
            week52 = pd.read_csv('datasets/monthly/{}'.format(filename))
            week52 = week52[(week52[['Volume']] != 0).all(axis=1)]
            name = filename.replace('.csv','')
            week52.to_csv('C:/Users/admin/Desktop/option chain analysis/datasets/monthly/{}.csv'.format(name))
            prev_avg = week52['Volume'].iloc[0:-1].mean()
            last_traded_price = week52['Volume'].iloc[-1]
            prev_avg = int(prev_avg)
            last_traded_price = last_traded_price.item()
            if last_traded_price > prev_avg:
                output.append(filename)
    else:
        print('do 1 nothing')
    return output
```

```python
def more_accuracy(stock,Nse_data):
    accu_out = []
    if Nse_data == 'Nifty50':
        for stockname in stock:
            rate = pd.read_csv('datasets/daily/{}'.format(stockname))
            if (rate['Close'].iloc[-1]).item() > (rate['Open'].iloc[-1]).item() and (rate['Close'].iloc[-2]).item() > (rate['Open'].iloc[-2]).item() and (rate['Close'].iloc[-3]).item() > (rate['Open'].iloc[-3]).item():
                accu_out.append(stockname)
            else:
                continue
    elif Nse_data == 'Nifty200':
        for stockname in stock:
            rate = pd.read_csv('datasets/monthly/{}'.format(stockname))
            if (rate['Close'].iloc[-1]).item() > (rate['Open'].iloc[-1]).item() and (rate['Close'].iloc[-2]).item() > (rate['Open'].iloc[-2]).item() and (rate['Close'].iloc[-3]).item() > (rate['Open'].iloc[-3]).item():
                accu_out.append(stockname)
            else:
                continue
    else:
        print('do nothing')
    return accu_out
```

```python
if __name__ == '__main__':
    Nse_data = str(input('Enter Nifty50 or Nifty200 : '))
    if Nse_data == 'Nifty50':
        Nifty_data = get_50nsedata()
    elif Nse_data == 'Nifty200':
        Nifty_data = get_200nsedata()
    else:
        print('invalid input')

    days = int(input('start date = Enter the no of days before current day : '))
    startdate = datetime.today() - timedelta(days)
    enddate = datetime.today()
    interval = str(input('Enter the Time interval like: 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk : ')) 

    #while True:
    i = 1
    while (i<=1):
        Hist_data = get_histdata(Nifty_data, startdate, enddate, interval,Nse_data)
        out = get_volume_analysis(Nse_data)
        print('output is : ', out)
        accuracy = str(input('Requierd more accuracy : Yes or No : '))
        if accuracy == 'Yes':
            cal = accu = more_accuracy(out,Nse_data)
            print('more accurate is : ', cal)
        else:
            pass
        i = i+1
        time.sleep(120)
```
